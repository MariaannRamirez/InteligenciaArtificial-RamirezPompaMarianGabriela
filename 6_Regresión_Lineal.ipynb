```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# 2. Generar datos simulados
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
# 3. Visualizar los datos
plt.scatter(X, y, color='blue')
plt.title('Datos simulados')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
# 4. Crear modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(X, y)
```

```python
# 5. Mostrar parámetros aprendidos
print(f"Intercepción (b): {modelo.intercept_[0]:.2f}")
print(f"Coeficiente (m): {modelo.coef_[0][0]:.2f}")
```

```python
# 6. Hacer predicciones
y_pred = modelo.predict(X)
```

```python
# 7. Visualizar línea de regresión
plt.scatter(X, y, color='blue', label='Datos reales')
plt.plot(X, y_pred, color='red', label='Regresión lineal')
plt.title('Regresión Lineal Ajustada')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# 8. Evaluar el modelo
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
print(f"Error cuadrático medio (MSE): {mse:.2f}")
print(f"Coeficiente de determinación (R²): {r2:.2f}")
```
